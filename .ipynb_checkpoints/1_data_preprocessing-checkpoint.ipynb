```python
df_train = pd.read_csv('traindata.csv')
df_train.info()
```

## 2. Explore the distribution of each variable

Plot categorical variable and numerical variable in the same figure

```python
import seaborn as sns
df_train = df_train.set_index('id')
X_train = df_train.drop(['stroke'], 1)

fig, axes = plt.subplots(len(df_train.drop('stroke',1).columns)//3, 3, figsize=(30, 20))
for col, axis in zip(df_train.drop('stroke',1).columns, axes.flatten()):
    df_temp = df_train[[col,'stroke']].dropna()
    if isinstance(df_temp[col].values[0], str):
        sns.countplot(x=col, hue="stroke", data=df_temp, ax = axis)
        axis.set_title(col)
    else:
        stroke_ = df_temp[df_temp['stroke']>0]
        nonstroke_ = df_temp[df_temp['stroke']==0]
        for a in [stroke_, nonstroke_]:
            sns.distplot(a[col].values, ax=axis, kde=False)
        axis.set_title(col)
plt.show()
```

### Conclusions from visualization results:
1. the number of age=0 is too high, should this be a default value when age is missing?
2. Gender = other?
3. Missing values in "smoking status" and "bmi"

### MICE imputation

```python
import fancyimpute
# why I have to use pip install fancyimpute instead of pip3 install fancy impute?
X_imputed = fancyimpute.MICE().complete(one_hot_encoded_training_predictors.as_matrix())
```

__Check imputed dataframe__

```python
imputed_predictors = pd.DataFrame(data = X_imputed, columns=one_hot_encoded_training_predictors.columns)
imputed_predictors.head(5)
```

## 5. Upsampling "stroke" samples
__Unbalanced data: we need to downsample non-stroke observations or upsample stroke observations___
* SMOTE
* SMOTEENN

```python
# using SMOTE to upsample stroke 
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTEENN
y_train = df_train['stroke']
sm1 = SMOTE(kind='regular',ratio={1:783*10}) 
X_smote, y_smote = sm1.fit_sample(imputed_predictors, y_train)

sm2 = SMOTEENN() 
X_smoteenn, y_smoteenn = sm2.fit_sample(imputed_predictors, y_train)
```

```python
print('training data size after upsampling is:',len(y_smoteenn))
```

```python
X_train_SMOTE= pd.DataFrame(X_smote, columns=imputed_predictors.columns)
X_train_SMOTEENN= pd.DataFrame(X_smoteenn, columns=imputed_predictors.columns)
```

```python
y_train_SMOTE= pd.DataFrame(y_smote, columns= ['stroke'])
y_train_SMOTEENN= pd.DataFrame(y_smoteenn, columns= ['stroke'])
```

```python
X_train_SMOTE.to_csv('X_train_resampled.csv',index = False)
y_train_SMOTE.to_csv('y_train_resampled.csv',index = False)

X_train_SMOTEENN.to_csv('X_train_smoteenn.csv',index = False)
y_train_SMOTEENN.to_csv('y_train_smoteenn.csv',index = False)
```
